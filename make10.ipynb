```julia
using Combinatorics
using DataFrames
```

```julia
num = collect(0:9)
```

重複組み合わせ

$_{n}H_{r} = \frac{(n+r-1)!}{r!(n-1)!}$

```julia
all_pattern = collect(with_replacement_combinations(num, 4))
```

```julia
binary_operator = [+, -, *, /, %, ^, log]
```

```julia
do_nothing(num::Int64) = num
unary_operator = [factorial, do_nothing]
```

```julia
for nums in all_pattern
    a,b,c,d = nums
    for bop1 in binary_operator
        for bop2 in binary_operator
            for bop3 in binary_operator
                for uop1 in unary_operator
                    for uop2 in unary_operator
                        for uop3 in unary_operator
                            for uop4 in unary_operator
                                result = bop3(bop2(bop1(uop4(a), uop3(b)), uop2(c)), uop1(d))
                                if result == 10
                                    println(a,b,c,d)
                                end
                            end
                        end
                    end
                end
            end
        end
    end
end
```

```julia
function my_div(a, b)
    if b == 0
        return Inf
    else
        return a/b
    end
end
```

```julia
function my_rem(a, b)
    if b == 0
        return NaN
    else
        return a%b
    end
end
```

```julia
binary_operator = [+, -, *, my_div, my_rem, ^]
```

```julia
for nums in all_pattern
    a,b,c,d = nums
    for bop1 in binary_operator
        for bop2 in binary_operator
            for bop3 in binary_operator
                for uop1 in unary_operator
                    for uop2 in unary_operator
                        for uop3 in unary_operator
                            for uop4 in unary_operator
                                result = bop3(bop2(bop1(uop4(a), uop3(b)), uop2(c)), uop1(d))
                                if result == 10
                                    println(a,b,c,d)
                                end
                            end
                        end
                    end
                end
            end
        end
    end
end
```
